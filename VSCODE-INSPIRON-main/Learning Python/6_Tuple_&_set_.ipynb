# **Set uses hashing**

Among `tuple`, `list`, and `set` in Python:

### ‚úÖ `set` uses the **concept of hashing**.

Here's a breakdown:

| Data Type | Mutable | Hashable                           | Uses Hashing   |
| --------- | ------- | ---------------------------------- | -------------- |
| `list`    | Yes     | No                                 | ‚ùå No           |
| `tuple`   | No      | Yes (if all elements are hashable) | ‚ùå Not directly |
| `set`     | Yes     | No                                 | ‚úÖ **Yes**      |

---

### üß† Why `set` uses hashing:

* A `set` is an **unordered collection of unique elements**.
* Internally, it uses a **hash table** for fast membership testing (`in`), insertion, and deletion.
* Only **hashable (immutable)** objects can be added to a set.

---

### üìå Bonus:

* `dict` in Python also uses **hashing** (keys must be hashable).
* You can check if an object is hashable using:

```python
hash(42)         # ‚úÖ Works
hash((1, 2))     # ‚úÖ Works
hash([1, 2])     # ‚ùå TypeError
```

Python **bitwise operators** operate on integers at the **binary level** (i.e., bit by bit). Here's a breakdown of all the bitwise operators with examples:

---

### ‚úÖ Bitwise Operators in Python

| Operator | Name                 | Description                                                | Example (`a = 10`, `b = 4`)                 |     |        |          |
| -------- | -------------------- | ---------------------------------------------------------- | ------------------------------------------- | --- | ------ | -------- |
| `&`      | AND                  | Sets each bit to 1 if both bits are 1                      | `a & b ‚Üí 10 & 4 = 0`                        |     |        |          |
| \`       | \`                   | OR                                                         | Sets each bit to 1 if at least one bit is 1 | \`a | b ‚Üí 10 | 4 = 14\` |
| `^`      | XOR                  | Sets each bit to 1 if bits are different                   | `a ^ b ‚Üí 10 ^ 4 = 14`                       |     |        |          |
| `~`      | NOT (1's complement) | Inverts all the bits (i.e., flips 0s to 1s and vice versa) | `~a ‚Üí ~10 = -11`                            |     |        |          |
| `<<`     | Left Shift           | Shifts bits to the left (adds zeros to the right)          | `a << 1 ‚Üí 10 << 1 = 20`                     |     |        |          |
| `>>`     | Right Shift          | Shifts bits to the right (drops bits from the right)       | `a >> 1 ‚Üí 10 >> 1 = 5`                      |     |        |          |

---

### üîç Binary Representation

To understand better:

```python
a = 10        # Binary: 1010
b = 4         # Binary: 0100
```

---

### üìò Examples

#### 1. AND `&`

```python
print(10 & 4)   # 1010 & 0100 = 0000 ‚Üí 0
```

#### 2. OR `|`

```python
print(10 | 4)   # 1010 | 0100 = 1110 ‚Üí 14
```

#### 3. XOR `^`

```python
print(10 ^ 4)   # 1010 ^ 0100 = 1110 ‚Üí 14
```

#### 4. NOT `~`

```python
print(~10)      # ~1010 = -(1010 + 1) = -11
```

#### 5. Left Shift `<<`

```python
print(10 << 1)  # 1010 << 1 = 10100 ‚Üí 20
```

#### 6. Right Shift `>>`

```python
print(10 >> 1)  # 1010 >> 1 = 0101 ‚Üí 5
```

---

### üí° Quick Tips

* `~x` gives `-(x+1)` because Python uses **two's complement** for negative binary numbers.
* Shifting left multiplies the number by `2^n`, shifting right divides it by `2^n`.

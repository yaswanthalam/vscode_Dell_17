In Python, **number systems** refer to the various ways numbers can be represented and handled‚Äîespecially in different bases. Python supports **decimal, binary, octal, and hexadecimal** number systems. Here's a breakdown:

---

### üî¢ 1. **Decimal (Base 10)**

* This is the default number system.
* Digits: `0-9`

```python
x = 25       # Decimal
print(x)     # Output: 25
```

---

### ‚ö´ 2. **Binary (Base 2)**

* Prefix: `0b` or `0B`
* Digits: `0 and 1`

```python
x = 0b1101
print(x)       # Output: 13
print(bin(x))  # Output: '0b1101'
```

---

### üü† 3. **Octal (Base 8)**

* Prefix: `0o` or `0O`
* Digits: `0-7`

```python
x = 0o17
print(x)       # Output: 15
print(oct(x))  # Output: '0o17'
```

---

### üî∂ 4. **Hexadecimal (Base 16)**

* Prefix: `0x` or `0X`
* Digits: `0-9` and `A-F` (or `a-f`)

```python
x = 0x1A
print(x)        # Output: 26
print(hex(x))   # Output: '0x1a'
```

---

### üîÅ Conversion Between Number Systems

Python provides built-in functions:

* `bin()` ‚Üí Converts to binary
* `oct()` ‚Üí Converts to octal
* `hex()` ‚Üí Converts to hexadecimal
* `int()` ‚Üí Converts from any base to decimal

```python
bin(10)          # '0b1010'
oct(10)          # '0o12'
hex(10)          # '0xa'
int('1010', 2)   # 10 (binary to decimal)
int('0xa', 16)   # 10 (hex to decimal)
```

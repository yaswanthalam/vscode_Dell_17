## ðŸ”¹ `if` Statement

The `if` statement is used to test a condition.
If the condition is **True**, the indented block of code runs. If itâ€™s **False**, Python skips it.

**Syntax:**

```python
if condition:
    # code block runs if condition is True
```

**Example:**

```python
x = 10

if x > 5:
    print("x is greater than 5")
```

âœ… Output:

```
x is greater than 5
```

---

## ðŸ”¹ `if-else` Statement

The `else` part runs only if the `if` condition is **False**.

**Syntax:**

```python
if condition:
    # code block runs if condition is True
else:
    # code block runs if condition is False
```

**Example:**

```python
x = 2

if x > 5:
    print("x is greater than 5")
else:
    print("x is not greater than 5")
```

âœ… Output:

```
x is not greater than 5
```

---

## ðŸ”¹ `if-elif-else` (Multiple Conditions)

Use `elif` (else if) when you want to check **more than one condition**.

**Example:**

```python
x = 0

if x > 0:
    print("Positive number")
elif x == 0:
    print("Zero")
else:
    print("Negative number")
```

âœ… Output:

```
Zero
```

* Indentation is **very important** in Python (usually 4 spaces).

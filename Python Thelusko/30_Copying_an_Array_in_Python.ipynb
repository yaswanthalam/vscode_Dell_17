# **The below code is shallow copy. In the shallow copy, we use VIEW function and then the values are changed to both the array functions. **


```python
from numpy import *

arr1=array([1,2,4,10,11])

arr2=arr1.view()

arr1[1]=7

print(arr1)
print(arr2)


print(id(arr1))

print(id(arr2))
```

# **The below one is DEEP COPY. In the deep copy the values are changed only in the arr1 and not on the other arrays.**

```python
from numpy import *

arr1=array([1,2,4,10,11])

arr2=arr1.copy()

arr1[1]=7

print(arr1)
print(arr2)


print(id(arr1))

print(id(arr2))
```

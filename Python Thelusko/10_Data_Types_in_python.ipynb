In Python, data types are categorized into **built-in types**. Here's a structured breakdown of the **main data types**:

---

### üî¢ **Numeric Types**

| Type      | Example  | Description                        |
| --------- | -------- | ---------------------------------- |
| `int`     | `42`     | Integer numbers                    |
| `float`   | `3.14`   | Floating-point numbers             |
| `complex` | `3 + 4j` | Complex numbers (real + imaginary) |

---

### üìö **Sequence Types**

| Type    | Example        | Description                   |
| ------- | -------------- | ----------------------------- |
| `str`   | `"hello"`      | Immutable text string         |
| `list`  | `[1, 2, 3]`    | Mutable ordered collection    |
| `tuple` | `(1, 2, 3)`    | Immutable ordered collection  |
| `range` | `range(0, 10)` | Immutable sequence of numbers |

---

### üî¢ **Set Types**

| Type        | Example              | Description              |
| ----------- | -------------------- | ------------------------ |
| `set`       | `{1, 2, 3}`          | Unordered, no duplicates |
| `frozenset` | `frozenset([1,2,3])` | Immutable version of set |

---

### üîë **Mapping Type**

| Type   | Example                        | Description     |
| ------ | ------------------------------ | --------------- |
| `dict` | `{"name": "Alice", "age": 30}` | Key-value pairs |

---

### üü¢ **Boolean Type**

| Type   | Example         | Description    |
| ------ | --------------- | -------------- |
| `bool` | `True`, `False` | Logical values |

---

### ‚ùå **None Type**

| Type       | Example | Description         |
| ---------- | ------- | ------------------- |
| `NoneType` | `None`  | Represents no value |

---

### üì¶ **Binary Types**

| Type         | Example                   | Description                   |
| ------------ | ------------------------- | ----------------------------- |
| `bytes`      | `b'hello'`                | Immutable byte sequence       |
| `bytearray`  | `bytearray([65, 66, 67])` | Mutable byte sequence         |
| `memoryview` | `memoryview(b'hello')`    | View of bytes without copying |

---

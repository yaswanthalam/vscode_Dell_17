## üîπ What is a `for` loop?

A `for` loop in Python is used to **iterate (loop through) a sequence** such as a list, tuple, string, dictionary, or a range of numbers.

---

## 1. Basic `for` loop

```python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)
```

‚úÖ Output:

```
apple
banana
cherry
```

üëâ The loop goes through each element of the list and executes the code.

---

## 2. Using `range()` with `for`

`range()` generates a sequence of numbers (default starts from 0).

```python
for i in range(5):
    print(i)
```

‚úÖ Output:

```
0
1
2
3
4
```

üëâ Here, `range(5)` means numbers from `0` to `4` (not including 5).

---

## 3. Range with start and step

```python
for i in range(2, 10, 2):  # start=2, stop=10, step=2
    print(i)
```

‚úÖ Output:

```
2
4
6
8
```

---

## 4. Looping through a string

```python
for letter in "Python":
    print(letter)
```

‚úÖ Output:

```
P
y
t
h
o
n
```

---

## 5. Using `for` with `else`

The `else` part runs **after the loop finishes normally** (without `break`).

```python
for i in range(3):
    print(i)
else:
    print("Loop finished")
```

‚úÖ Output:

```
0
1
2
Loop finished
```

---

## 6. Nested `for` loops

```python
for i in range(3):
    for j in range(2):
        print(i, j)
```

‚úÖ Output:

```
0 0
0 1
1 0
1 1
2 0
2 1
```

---

‚ö° Key points:

* `for` loops are used for iterating over sequences.
* `range()` is very common with loops.
* Indentation is **required** in Python.

# **Left Shift operator <<**


Example:

5 << 2

5 in binary: 00000101

Shifted 2 positions left: 00010100

00010100 in decimal: 20

Effect:

Equivalent to multiplying the first operand by 2 raised to the power of the second operand.

5 << 2 is the same as 5 * 2^2 (5 * 4 = 20)

```python
10<<2
```

```python
52225552<<8555
```

```python
2<<55
```

# **Right shift operator >> **


Example:

20 >> 2

20 in binary: 00010100

Shifted 2 positions right: 00000101

00000101 in decimal: 5

Effect:

Equivalent to dividing the first operand by 2 raised to the power of the second operand (integer division, discarding any remainder).

20 >> 2 is the same as 20 / 2^2 (20 / 4 = 5)



```python
20>>2
```

# **Check whether a number is Prime or Not ?**

```python
import math

def is_prime(num):
    """
    Checks if a number is prime.

    Args:
        num (int): The number to check.

    Returns:
        bool: True if the number is prime, False otherwise.
    """
    if num < 2:
        return False

    for i in range(2, int(math.sqrt(num)) + 1):
        if num % i == 0:
            return False

    return True

# Example usage:
number = 10
if is_prime(number):
    print(f"{number} is a prime number")
else:
    print(f"{number} is not a prime number")
```

```python
import socket
import time

target = "8.8.8.8"
port = 443

sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
sock.settimeout(1)

start_time = time.time()
sock.sendto(b"ping", (target, port))
try:
    data, addr = sock.recvfrom(1024)
    print(f"Response from {addr}: {data.decode()}")
    print(f"Time: {time.time() - start_time} seconds")
except socket.timeout:
    print("No response received")
```

```python
from netmiko import ConnectHandler
from datetime import datetime

devices = [
    {'device_type': 'cisco_ios', 'host': '192.168.1.1', 'username': 'admin', 'password': 'password123'},
    # Add more devices as needed
]

for device in devices:
    connection = ConnectHandler(**device)
    output = connection.send_command('show running-config')

    # Create a filename based on the device hostname and current timestamp
    filename = f"{device['host']}_backup_{datetime.now().strftime('%Y%m%d%H%M%S')}.txt"

    with open(filename, 'w') as backup_file:
        backup_file.write(output)

    print(f"Backup for {device['host']} saved as {filename}")
    connection.disconnect()
```

```python
import math

# Square Root
x = math.sqrt(25)  # Returns the square root of 25, which is 5.0
print(f"Square root of 25: {x}")

# Power
y = math.pow(2, 3)  # Returns 2 raised to the power of 3, which is 8.0
print(f"2 raised to the power of 3: {y}")

# Trigonometric Functions (sin, cos, tan)
angle_degrees = 45
angle_radians = math.radians(angle_degrees)  # Convert degrees to radians

sin_value = math.sin(angle_radians)
cos_value = math.cos(angle_radians)
tan_value = math.tan(angle_radians)

print(f"Sine of 45 degrees: {sin_value}")
print(f"Cosine of 45 degrees: {cos_value}")
print(f"Tangent of 45 degrees: {tan_value}")

# Ceiling and Floor
num = 3.7
ceil_value = math.ceil(num)  # Returns the smallest integer greater than or equal to num (4)
floor_value = math.floor(num)  # Returns the largest integer less than or equal to num (3)

print(f"Ceiling of 3.7: {ceil_value}")
print(f"Floor of 3.7: {floor_value}")

# Absolute Value
abs_value = math.fabs(-5)  # Returns the absolute value of -5, which is 5.0
print(f"Absolute value of -5: {abs_value}")

# Logarithms
log_value = math.log(100, 10)  # Returns the base-10 logarithm of 100, which is 2.0
print(f"Base-10 logarithm of 100: {log_value}")

# Constants
pi_value = math.pi  # Returns the value of pi (approximately 3.14159)
e_value = math.e  # Returns the value of Euler's number (approximately 2.71828)

print(f"Value of pi: {pi_value}")
print(f"Value of e: {e_value}")
```

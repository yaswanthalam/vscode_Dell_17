# ** Break, Continue and PASS**

# **Break **

```python
av=10
x=int(input("How many candies you want: "))

i=1
while i<=x:
  if i>av:
    print("The maximum limit is 10")
    break
  print('candy')
  i+=1
```

# **Continue**

```python
for i in range(1,10):
  if i%3==0:
    continue
  print(i)
print("bye")
```

```python
for i in range(10):
  if i%3==0:
    continue
  print(i)
print("bye")
```

```python
for i in range(1,10):
  if i%3==0 or i%5==0:
    continue
  print(i)
print("bye")
```

# ***Pass***

```python
for i in range(1,10):
  if i%2!=0:
    pass
  else:
    print(i)
print("bye")
```

```python
for i in range(1,10):
  if i%2==0:
    break
  else:
    print(i)
print("bye")
```

```python
for i in range(1,10):
  if i%2==0:
    continue
  else:
    print(i)
print("bye")
```

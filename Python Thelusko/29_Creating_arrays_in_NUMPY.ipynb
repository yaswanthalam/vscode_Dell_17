# **LINSPACE**

The Linspace will break the range into multiple parts.

For example: 1,5,10.

from the numbers 1 to 5 it will divide into 10 parts. Dividing like this 10 parts is done with the help of linspace

```python
from numpy import *

arr = linspace(0,15,16)

print(arr)
```

```python
from numpy import *

arr = linspace(0,15,16)

print(arr)
```

```python
from numpy import *

arr = linspace(0,1,2)

print(arr)
```

```python
from numpy import *

arr = linspace(0,10,10)

print(arr)
```

```python
from numpy import *

arr = linspace(0,10,20)

print(arr)
```

```python
from numpy import *

arr = linspace(0,15,20)

print(arr)
```

```python
from numpy import *

arr = linspace(0,15,50)

print(arr)
```

# **Arrange**

# **ZEROS()  and Ones()**

```python
from numpy import *

arr = ones(10)

print(arr)
```

```python
from numpy import *

arr = zeros(10)

print(arr)
```

```python
from numpy import *

arr = zeros(10, int)

print(arr)
```

```python
from numpy import *

arr = ones(10, int)

print(arr)
```

```python
from numpy import *

arr = ones(100, int)

print(arr)
```

```python
from numpy import *

arr = ones(100, float)

print(arr)
```

```python
from numpy import *

arr = ones(100, bool)

print(arr)
```

```python
from numpy import *

arr = zeros(100, bool)

print(arr)
```

# ***Position***

```python
def sum(a,b):
  c=a+b
  print(c)

sum(6,5)

```

```python
def person(name, age):
  print(name)
  print(age)

person('navin',28)
```

```python
def person(name, age):
  print(name)
  print(age)

person('navin',age=28)
```

# **Key Word**

```python
def person(name, age):
  print(name)
  print(age)

person(age=28, name='navin')
```

```python
def person(name, age):
  print(name)
  print(age)

person(28,'navin')
```

# **Default**

```python
 def person(name, age=28):
  print(name)
  print(age)

person('navin')
```

```python
def person(name, age=28):
  print(name)
  print(age)

person('navin', 26)
```

# **Variable Length**

```python
def sum(a,b):
  c=a+b
  print(c)

sum(5,6)
```

```python
def sum(a,*b):
  c=a
  for i in b:
    c=c+i
  print(c)

sum(5,6,85,96,522,2,5,3,5,6)
```

```python
def sum(a,*b):
  c=a
  for i in b:
    c=c+i
    print(c)

sum(5,6,85,96,522,2,5,3,5,6)
```

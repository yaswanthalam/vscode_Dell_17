# Keyworded Variable Length Arguments in Python | **kwargs

image.png

```python
def person(name, **data):
  print(name)
  print(data)


person('navin',age=28, city='mumbai', mob=892855)
```

```python
def person(name, **data):
  print(name)
  for i,j in data.items():
    print(i,j)


person('navin',age=28, city='mumbai', mob=892855)
```

```python
def yaswanth(name, **data):
  print(name)

  for i,j in data.items():
    print(i,j)

yaswanth("Yaalam" , age = 522 , city = "TPT" , mob = 4551515215455)
```

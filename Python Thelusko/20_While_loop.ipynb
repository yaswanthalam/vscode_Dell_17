In Python, a `while` loop is used to repeatedly execute a block of code **as long as a given condition is `True`**.

---

### üîÅ Syntax of a `while` loop:

```python
while condition:
    # block of code to execute
```

* The `condition` is evaluated before each iteration.
* The loop will keep running until the `condition` becomes `False`.

---

### ‚úÖ Example 1: Basic while loop

```python
count = 1

while count <= 5:
    print("Count is:", count)
    count += 1
```

**Output:**

```
Count is: 1
Count is: 2
Count is: 3
Count is: 4
Count is: 5
```

---

### üîÅ Infinite loop (be careful!)

```python
while True:
    print("This will run forever unless you break it")
```

To stop it, you‚Äôll usually use:

```python
break
```

---

### üîÑ Example 2: `while` with `break`

```python
x = 1
while True:
    print(x)
    x += 1
    if x > 3:
        break
```

**Output:**

```
1
2
3
```

---

### üîÅ Example 3: `while` with `continue`

```python
x = 0
while x < 5:
    x += 1
    if x == 3:
        continue
    print(x)
```

**Output:**

```
1
2
4
5
```

(`3` is skipped because of `continue`)

# **Using While Loop**

```python
from array import *
vals = array('i',[5,-9,4,2])

newArr = array(vals.typecode , (a for a in vals))

i=0

while i<len(newArr):
  print(newArr[i])
  i=i+1


```

```python
from array import *
vals = array('i',[5,-9,4,2])

newArr = array(vals.typecode , (a*a for a in vals))

i=0

while i<len(newArr):
  print(newArr[i])
  i=i+1
```

# **The below code will not swap the variables instead it will reassign the variables **

```python
a=5
b=10

a=b
b=a

print(a,b)
```

```python
a=5
b=10

c=a
a=b
b=c

print(a,b)
```

```python
a=5
b=10

a=a+b
b=a-b
a=a-b

print(a,b)
```

```python
a=5
b=10

a=a^b
b=a^b
a=a^b

print(a,b)
```

```python
a=5
b=10

a,b=b,a

print(a,b)
```

# **Strings in Python are immutable**

```python
hello[0:3]='my'
```

The error TypeError: 'str' object does not support item assignment occurs because strings in Python are immutable. This means you cannot change individual characters or slices of an existing string.

To achieve a similar result, you need to create a new string by concatenating the desired parts. Here's how you can replace the first three characters of the hello string:

```python
hello[0]='R'
```

```python
'my'+hello[3:6]
```

```python
hello
```

```python
'yaswanth'+ hello[0:10]
```

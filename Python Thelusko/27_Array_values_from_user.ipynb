```python
from array import *
arr = array('i',[])

n = int(input("Enter the lenth of ARRAY :  "))

for i in range(n):
  x = int(input("Enter the next Value :  "))
  arr.append(x)

print(arr)

val = int(input("Enter the value for search "))

k=0

for e in arr:
  if e==val:
    print(k)
    break
  k=k+1

print(arr.index(val))
```

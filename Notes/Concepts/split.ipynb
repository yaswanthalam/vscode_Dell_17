## What is `split()` in Python?

* `split()` is a **string method** in Python.
* It **splits a string into a list of substrings** based on a specified delimiter (separator).
* By default, it splits on **any whitespace** (spaces, tabs, newlines).
* It returns a **list** of the parts after splitting.

---

## Syntax

```python
str.split(separator=None, maxsplit=-1)
```

* `separator` (optional): The delimiter string where the split happens. If `None` (default), splits on any whitespace.
* `maxsplit` (optional): Maximum number of splits to do. Default `-1` means "no limit" (split all occurrences).

---

## How `split()` works?

* It scans the string.
* Whenever it finds the `separator`, it breaks the string there.
* Returns a list of all substrings between separators.
* If no separator is found, returns a list with the whole string as one element.

---

## Use Cases

* **Tokenizing words in a sentence** (e.g., for text processing).
* Splitting **CSV or TSV data** lines into fields.
* Parsing **logs** or **user input** separated by spaces or other delimiters.
* Extracting parts of data like **email username/domain** by splitting on `@`.
* Breaking strings into manageable chunks to process.

---

## Examples

### 1. Basic split by whitespace

```python
sentence = "Hello world, welcome to Python!"
words = sentence.split()
print(words)
# Output: ['Hello', 'world,', 'welcome', 'to', 'Python!']
```

### 2. Split by a specific character (e.g., comma)

```python
data = "apple,banana,cherry,dates"
fruits = data.split(',')
print(fruits)
# Output: ['apple', 'banana', 'cherry', 'dates']
```

### 3. Using `maxsplit` to limit splits

```python
text = "one two three four five"
result = text.split(' ', 2)
print(result)
# Output: ['one', 'two', 'three four five']
```

### 4. Splitting an email to get username and domain

```python
email = "user123@example.com"
username = email.split('@')[0]
domain = email.split('@')[1]

print("Username:", username)  # Output: user123
print("Domain:", domain)      # Output: example.com
```

### 5. Split a path

```python
path = "/home/user/documents/file.txt"
folders = path.split('/')
print(folders)
# Output: ['', 'home', 'user', 'documents', 'file.txt']
```

---

## Important Notes

* If the separator is not found in the string, you get a list with the entire string as the only element.
* Leading and trailing separators result in empty strings in the output list.
* Splitting on whitespace treats consecutive whitespace as a single separator, so no empty strings in that case.

### Example 1: Counting words in a sentence

```python
sentence = input("Enter a sentence: ")  # User inputs: Hello world welcome to Python

words = sentence.split()  # split on whitespace
word_count = len(words)

print(f"The sentence has {word_count} words.")
```

**What happens:**
User inputs a sentence, we split it into words by spaces, then count how many words there are.

---

### Example 2: Parse CSV line into fields

Imagine you get a CSV line from a file and want to separate fields:

```python
csv_line = "John,Doe,30,New York"
fields = csv_line.split(',')

print("First name:", fields[0])
print("Last name:", fields[1])
print("Age:", fields[2])
print("City:", fields[3])
```

---

### Example 3: Extract username from email addresses entered by user

```python
emails = ["alice@example.com", "bob@domain.org", "carol@mail.com"]

for email in emails:
    username = email.split('@')[0]
    print(f"Username for {email} is {username}")
```

---

### Example 4: Limit splits for fixed-format strings

Say you have a log entry with date, time, level, and message separated by spaces, but message can contain spaces:

```python
log = "2025-12-25 11:00:00 INFO System rebooted successfully"

parts = log.split(' ', 3)
date = parts[0]
time = parts[1]
level = parts[2]
message = parts[3]

print(f"Date: {date}, Time: {time}, Level: {level}, Message: {message}")
```

Here, `maxsplit=3` ensures only first three spaces split the string; everything else stays in the message.

---

### Example 5: Split user input into commands

If you are building a simple command-line app where user inputs commands and parameters:

```python
user_input = input("Enter command: ")  # e.g. "move north 10 steps"

tokens = user_input.split()

command = tokens[0]
args = tokens[1:]

print(f"Command: {command}")
print(f"Arguments: {args}")
```

### Summary

* `split()` makes it easy to **break strings into parts** based on delimiters.
* It’s widely used in **parsing input, processing data files, tokenizing text, or extracting information**.
* Using the optional parameters like `maxsplit` gives you **control** over how many splits to do.

Given a number 5, write a program to find its factorial.

Task 24: Find Maximum Number in a List (Without max())

Question:
Given the list [4, 7, 2, 9, 1], find the maximum number without using max().

Expected Output:

9



````python
lacp = [4, 7, 2, 9, 1]

print(max(lacp))
````

````python
lacp = [4, 7, 2, 9, 1]

lldp = lacp[0]

for i in lacp:
    if i > lldp:
        lldp = i

print(lldp)
````

````python
abcd = (1,5,8,9,4,10,52,63,20,12,30,78,96,45,20,15)

xyz = abcd[0]

for i in abcd:
    if i > xyz:
        xyz = i
print(xyz)
````

You‚Äôve written **three very similar programs**, but the **position of `print(lldp)`** changes the output and behavior. Let‚Äôs go through **each one step by step** so you can clearly see the difference.

---

## üîπ Code 1

```python
lacp = [4, 7, 2, 9, 1]

lldp = lacp[0]

for i in lacp:
    if i > lldp:
        lldp = i

print(lldp)
```

### ‚úÖ What this code does

* Finds the **maximum number in the list**
* Prints the result **once at the end**

### üîç Step-by-step execution

| Iteration | i | lldp (current max) | Condition |
| --------- | - | ------------------ | --------- |
| Start     | ‚Äî | 4                  | ‚Äî         |
| 1         | 4 | 4                  | 4 > 4 ‚ùå   |
| 2         | 7 | 7                  | 7 > 4 ‚úÖ   |
| 3         | 2 | 7                  | 2 > 7 ‚ùå   |
| 4         | 9 | 9                  | 9 > 7 ‚úÖ   |
| 5         | 1 | 9                  | 1 > 9 ‚ùå   |

### üñ® Output

```text

```

‚úîÔ∏è **Correct and recommended** way to find the maximum using a loop.

---

## üîπ Code 2

```python

    if i > lldp:

    print(lldp)
```

### ‚úÖ What this code does

Question:
Print numbers between 1 and 30 that are divisible by both 3 and 5.

Expected Output:

15
30




````python
for i in range(1, 31):
    if i % 3 == 0 and i % 5 == 0:
        print(i)
````

````python
hello = 30

hi = 1

for i in range(hi, hello+1):
    if (i % 3 == 0 and i % 5 == 0):
        print(i)
````

Almost there üëç ‚Äî but it is still not correct because of one key mistake.

‚ùå Problem

You are using & instead of and:

i % 3 == 0 & i % 5 == 0


& is a bitwise operator

and is the logical operator you need for conditions

Because of operator precedence, your condition does not work as intended.

````python
hello = 31
for i in range(hello): 
    if (i % 3 == 0 and i % 5 == 0):
         print(i)
````

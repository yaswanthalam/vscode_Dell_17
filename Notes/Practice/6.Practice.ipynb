Sure! I'll explain what each of the common Python methods for counting words **actually does** step-by-step, so you get the idea clearly.

---

### 1. `sentence.split()` — Basic word count by splitting on spaces

```python
sentence = "Python is very easy to learn"
words = sentence.split()  # Splits sentence into list of words by whitespace
word_count = len(words)   # Counts how many words are in the list
print(word_count)
```

**What happens here?**

* `split()` takes the whole sentence string and **breaks it into pieces ("words") whenever it sees whitespace** (spaces, tabs, newlines).
* It returns a **list** of words, e.g. `["Python", "is", "very", "easy", "to", "learn"]`.
* `len()` counts how many items are in that list.
* So, the result is the number of words separated by whitespace.

---

### 2. `re.findall(r'\b\w+\b', sentence)` — Using Regular Expressions to find words

```python
import re
sentence = "Python is easy, fast, and powerful!"
words = re.findall(r'\b\w+\b', sentence)
word_count = len(words)
print(word_count)
```

**What does this do?**

* `re.findall()` searches the string and **finds all substrings that match the pattern** given.
* The pattern `\b\w+\b` means:

  * `\w+` = one or more "word characters" (letters, digits, underscore)
  * `\b` = word boundary (start or end of a word)
* So this pattern matches all whole words, **ignoring punctuation like commas or exclamation marks**.
* It returns a list of matched words.
* `len()` then counts how many words were found.

This is better than `split()` when your sentence contains punctuation.

4.  
Question: Given `text = "hello world"`, print the same string but with each word capitalized.  
Expected output:  
`Hello World`


````python
text = "hello world"

print(text.upper())
````

You can achieve this in Python using the .title() method or by splitting the string into words, capitalizing each, and then joining them back. Here's a simple way:

````python
text1 = "hello world"

print(text1.title())
````

    Alternate way to split the code. 

````python
text = "hello world"
words = text.split()               # Split into words
capitalized_words = [w.capitalize() for w in words]  # Capitalize each word
result = " ".join(capitalized_words)  # Joins back with space
print(result)
````

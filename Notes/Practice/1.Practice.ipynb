string[start : end : step]


In your code:

exp[::-1]

Part	Meaning
start	empty  start from beginning
end	empty  go till end
step	-1  move backwards

 So Python reads the string from last character to first.

Example:

5 4 3 2 1 0    reversed

"nohtyp"

Task 18: Find Length of a List Without len()

Question:
Given the list [1, 2, 3, 4], find its length without using len().

Expected Output:

4

````python
vnet = [1,2,3,4]

count = 0

for i in vnet:
    count += 1

print(count)

````

````python
vnet = [1,2,3,4]

print(len(vnet))
````

````python
lst = [1, 2, 3, 4]

count = 0
for _ in lst:
    count += 1

print(count)
````

Heres a clear step-by-step explanation of the code:

```python
lst = [1, 2, 3, 4]
```

* This line creates a list named `lst` containing four elements: `1, 2, 3, 4`.

---

```python
count = 0
```

* A variable `count` is initialized to `0`.
* This will be used to keep track of how many elements are in the list.

---

```python
for _ in lst:
```

* This `for` loop goes through **each element** in the list `lst`.
* The underscore `_` is used because we **dont need the actual values** (1, 2, 3, 4), only the number of times the loop runs.

---

```python
    count += 1
```

* Each time the loop runs, `count` is increased by `1`.
* Since the loop runs once for every element in the list, `count` increases once per item.

---

```python
print(count)
```

* After the loop finishes, `count` contains the total number of elements in the list.
* The value is printed.

---

### Final Output:

```
4
```

 This method finds the length of the list **without using `len()`** by manually counting each element.


````python
lst = {1, 2, 3, 4}

count = 0
for _ in lst:
    count += 1

print(count)
````

Method 5: Using sum()

````python
cg1 = "banana"

print(sum(1 for ch in cg1 if ch == "a"))
````

Let‚Äôs break this **line by line**, very clearly and slowly üëç

### Code:

```python
cg1 = "banana"
print(sum(1 for ch in cg1 if ch == "a"))
```

---

## üîπ Step 1: Understand the string

```python
cg1 = "banana"
```

The string contains:

```
b  a  n  a  n  a
```

---

## üîπ Step 2: Look at the generator expression

```python
(1 for ch in cg1 if ch == "a")
```

This part means:

* `for ch in cg1` ‚Üí loop through **each character** in `"banana"`
* `if ch == "a"` ‚Üí **only keep** characters that are `"a"`
* `1` ‚Üí for **each match**, produce the value `1`

So internally, Python creates this sequence:

```
1, 1, 1
```

(one `1` for each `"a"` found)

---

## üîπ Step 3: What does `sum()` do?

```python
sum(1, 1, 1)
```

* `sum()` adds all the values together
* `1 + 1 + 1 = 3`

---

## üîπ Step 4: Final output

```python
print(3)
```

### ‚úÖ Output:

```
3
```

---

## üß† Think of it in simple English

> ‚ÄúGo through each character in `banana`.
> Every time you see an `'a'`, count **1**.
> Add all those 1‚Äôs together.‚Äù

---

## üîÅ Same logic written in long form (for comparison)

```python
count = 0
for ch in cg1:
    if ch == "a":
        count += 1

print(count)
```

Both versions do **exactly the same thing** ‚úÖ

---

### ‚≠ê Why this method is useful

* Very **Pythonic**
* Efficient (doesn‚Äôt create a list in memory)
* Common in **interviews**

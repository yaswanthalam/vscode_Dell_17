Great example üëç
This does the **same job** as list comprehension, but using **`filter()` + `lambda`**.
Let‚Äôs break it down **step by step in detail**.

---

## **Code**

```python
nums = [1, 2, 3, 2, 4, 2, 5]
value_to_remove = 2

result = list(filter(lambda x: x != value_to_remove, nums))
print(result)
```

---

## **Step 1: The list**

```python
nums = [1, 2, 3, 2, 4, 2, 5]
```

* Contains repeated values
* We want to remove **all occurrences of `2`**

---

## **Step 2: Value to remove**

```python
value_to_remove = 2
```

* This is the value we want to filter out

---

## **Step 3: Understanding `filter()`**

### **Syntax**

```python
filter(function, iterable)
```

* `function` ‚Üí must return **True or False**
* `iterable` ‚Üí list, tuple, etc.
* `filter()` keeps elements where the function returns **True**

‚ö†Ô∏è `filter()` returns a **filter object**, not a list ‚Üí so we convert it using `list()`.

---

## **Step 4: Understanding `lambda`**

```python
lambda x: x != value_to_remove
```

This is a **small anonymous function**, equivalent to:

```python
def check(x):
    return x != value_to_remove
```

---

## **Step 5: How `filter()` works internally**

Python goes through `nums` **one element at a time**:

| Element (`x`) | Check `x != 2` | Keep it? |
| ------------- | -------------- | -------- |
| 1             | True           | ‚úÖ Yes    |
| 2             | False          | ‚ùå No     |
| 3             | True           | ‚úÖ Yes    |
| 2             | False          | ‚ùå No     |
| 4             | True           | ‚úÖ Yes    |
| 2             | False          | ‚ùå No     |
| 5             | True           | ‚úÖ Yes    |

Only values returning **True** remain.

---

## **Step 6: Converting to list**

```python
list(filter(...))
```

* Converts the filter object into a list

Result:

```python
[1, 3, 4, 5]
```

## **Final Output**

```text
[1, 3, 4, 5]
```

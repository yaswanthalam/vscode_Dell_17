```python
from abc import ABC, abstractmethod

class computer:
    @abstractmethod
    #name 
    def process(self):
        #parameter that accepts arguments
        pass
        # decelaration and no definition are called as abstract methods. 

com = computer()
com.process()


```

```python
from abc import ABC, abstractmethod

class computer(ABC):
    @abstractmethod
    #name 
    def process(self):
        #parameter that accepts arguments
        pass
        # decelaration and no definition are called as abstract methods. 

class laptop(computer):
    def process(self):
        print("it's running Alam")

#com = computer()
com1 = laptop()
com1.process()


```

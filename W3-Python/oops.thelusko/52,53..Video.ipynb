```python
class student:

    school = "Thelusko"

    def __init__(self,m1,m2,m3):
        self.m1 = m1
        self.m2 = m2
        self.m3 = m3

    def avg(self):
        return(self.m1 + self.m2 +self.m3)
    
    def info(cls):
        return cls.school

s1 = student(84,5,15)
s2 = student(44,451,541541541)


print(s1.avg(), s2.avg())
print(student.info())
```

we need to pass the decorator @classmethod to call it correctly. 

```python
class student:

    school = "Thelusko"

    def __init__(self,m1,m2,m3):
        self.m1 = m1
        self.m2 = m2
        self.m3 = m3

    def avg(self):
        return(self.m1 + self.m2 +self.m3)
    
    @classmethod
    def info(cls):
        return cls.school

s1 = student(84,5,15)
s2 = student(44,451,541541541)


print(s1.avg(), s2.avg())
print(student.info())
```

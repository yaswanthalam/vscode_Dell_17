compare is not an inbuilt method, we have define it and write and function and work on it. 

```python
class comp:
    def __init__(self):
        self.name = "yaswanth"
        self.age = 51541515145
    
    def update(self):
        self.age = 96

    def compare(self, other):
        if self.age == other.age:
            return True
        else:
            return False

c1 = comp()
c2 = comp()

if c1.compare(c2):
    print("They are same")


c1.update()

print(c1.name)
print(c1.age)
print(c2.name)
print(c2.age)
```

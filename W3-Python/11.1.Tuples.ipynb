Python - Unpack Tuples

Unpacking a Tuple

When we create a tuple, we normally assign values to it. This is called "packing" a tuple:


```python
fruits = ("apple", "banana", "cherry")

print(fruits)
```

```python
abc = ("yaswanth" ,"alam" , "alam" , 41,41, "85", "85" , "4151515", 0,5451,"and","jk")

print(abc)
```

But, in Python, we are also allowed to extract the values back into variables. This is called "unpacking":

Example

Unpacking a tuple:

```python
fruits = ("apple", "banana", "cherry")

(green, yellow, red) = fruits

print(green)
print(yellow)
print(red)
```

Note: The number of variables must match the number of values in the tuple, if not, you must use an asterisk to collect the remaining values as a list.

```python
abc = ("alam" , "ab" , 545, "5451")

(alam, ab , xn , mn) = abc

print(alam)
print(ab)
print(xn)
print(mn)

print(alam, ab, xn, mn)
```

Using Asterisk*

If the number of variables is less than the number of values, you can add an * to the variable name and the values will be assigned to the variable as a list:

Example

Assign the rest of the values as a list called "red":

```python
fruits = ("apple", "banana", "cherry", "strawberry", "raspberry")

(green, yellow, *red) = fruits

print(green)
print(yellow)
print(red)
```

```python
yaswanth = ("alam" "sdfnjs" , "sdfghsdfj" ,"sfghdfjd" ,"sfgdjn" , "xcnvjx" , 5454154)

(al, nc, *guy) = yaswanth

print(al)
print(nc)
print(guy)
```

If the asterisk is added to another variable name than the last, Python will assign values to the variable until the number of values left matches the number of variables left.

Example


Add a list of values the "tropic" variable:

```python
fruits = ("apple", "mango", "papaya", "pineapple", "cherry")

(green, *tropic, red) = fruits

print(green)
print(tropic)
print(red)
```

```python
yas = ("ajsdnj" ,"xfgdfn" , 5156125 , 21020121, 8798959, "xnxjfndjk", "sdfhdfjn")

(abc , xyz , lon) = yas

print(abc)
print(xyz)
print(lon)
```

Python - Loop Tuples

Loop Through a Tuple


You can loop through the tuple items by using a for loop.


```python
abc = ("yas","mnc" ,"har")

for i in abc:
    print(abc)
```

```python
abc = ("yas","mnc" ,"har")

for i in abc:
    print(i)
```

Loop Through the Index Numbers

You can also loop through the tuple items by referring to their index number.

Use the range() and len() functions to create a suitable iterable.

```python
abc = ("yas","mnc" ,"har")

for i in range(len(abc)):
    print(abc[i])
```

```python
thistuple = ("apple", "banana", "cherry")
for i in range(len(thistuple)):
  print(thistuple[i])
```

```python
ji = ("asd" ,"asdfsd" ,"sfnbsdfj" ,"xfndf" ,"xffdnj",62165125, "545151")

for i in range(len(ji)):

    print(ji[i])
```

```python
jm = ("asd" ,"asdfsd" ,"sfnbsdfj" ,"xfndf" ,"xffdnj",62165125, "545151")

for i in range(len(jm)):
    print(jm[i])
```

Using a While Loop

You can loop through the tuple items by using a while loop.

Use the len() function to determine the length of the tuple, then start at 0 and loop your way through the tuple items by referring to their indexes.

Remember to increase the index by 1 after each iteration.

Example

Print all items, using a while loop to go through all the index numbers:

```python
jm = ("asd" ,"asdfsd" ,"sfnbsdfj" ,"xfndf" ,"xffdnj",62165125, "545151")

i = 0

while i < len(jm):
    print(jm[i])
    i = i+1
```

```python
bc = ("asd" ,"asdfsd" ,"sfnbsdfj" ,"xfndf" ,"xffdnj",62165125, "545151")

i = 0

while i < len(bc):
    print(bc[i])
    i = i +1
```

```python
mytuple = ('apple', 'banana', 'cherry')
i = 0
while i < len(mytuple):
  print(mytuple[i])
  i = i + 1
```

Python - Join Tuples


To join two or more tuples you can use the + operator:

```python
abc = ("sdf" , "xfjn" ,"sdfbsjf")
bcd = (55, 546, 6326, 5995)

cdm = abc + bcd

print(cdm)
```

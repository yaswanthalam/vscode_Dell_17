```python
class Person:
  species = "Human" # Class property

  def __init__(self, name):
    self.name = name # Instance property

p1 = Person("Emil")
p2 = Person("Tobias")

print(p1.name)
print(p2.name)
print(p1.species)
print(p2.species)
```

```python
class phase1:
    phase2 = "DPD time out and SA life time"  # class property

    def __init__(self, primary):
        self.primary = primary  # instance property


msee1 = phase1("PFS key")
msee2 = phase1("DPD time out")

print(msee1.primary)
print(msee2.primary)
print(msee1.phase2)
print(msee2.phase2)
```

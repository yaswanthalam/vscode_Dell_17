```python
class Calculator:
  def add(self, a, b):
    return a + b

  def multiply(self, a, b):
    return a * b

calc = Calculator()
print(calc.add(5, 3))
print(calc.multiply(4, 7))
```

```python
class alam:
    def add(self, a , b):
        return a + b
    
    def multiply(self, a , b):
        return a * b
    
calc = alam()
print(calc.add(5,10))
print(calc.multiply(2,4))
```

```python
class vpn:
    def add(self, x , y):
        return x + y

    def sub(self, x , y):
        return x - y

exp = vpn()
print(exp.add(5,8))
print(exp.sub(2,16))
```

```python
class nsg:
    def asg(self, t , i):
        return t + i
    def ddos(self, t , i):
        return t - i

cdn = nsg()
print(cdn.asg(5,10))
print(cdn.ddos(15,5))
```

```python
class Playlist:
  def __init__(self, name):
    self.name = name
    self.songs = []

  def add_song(self, song):
    self.songs.append(song)
    print(f"Added: {song}")

  def remove_song(self, song):
    if song in self.songs:
      self.songs.remove(song)
      print(f"Removed: {song}")

  def show_songs(self):
    print(f"Playlist '{self.name}':")
    for song in self.songs:
      print(f"- {song}")

my_playlist = Playlist("Favorites")
my_playlist.add_song("Bohemian Rhapsody")
my_playlist.add_song("Stairway to Heaven")
my_playlist.show_songs()
```

```python
class vnet:
    def __init__(self, subnet):
        self.subnet = subnet 
        self.nsgs = []
    def route_table(self, nsg):
        self.nsgs.append(nsg)
        print(f"Security Group: {nsg}")
    def peering(self, nsg):
        self.nsgs.remove(nsg)
        print(f"Application security group: {nsg}")
    def ddos_plan(self):
        print(f"IP based plan and the network based plan '{self.subnet}' :")
        for security in self.nsgs:
            print(f"-{security}")

azure = vnet("Management Groups")
azure.route_table("SEcurity table")
azure.route_table("DDOS and Bastion groups")
azure.ddos_plan()
```

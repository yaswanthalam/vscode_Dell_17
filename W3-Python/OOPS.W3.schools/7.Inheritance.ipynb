```python
class Person:
  def __init__(self, fname, lname):
    self.firstname = fname
    self.lastname = lname

  def printname(self):
    print(self.firstname, self.lastname)

#Use the Person class to create an object, and then execute the printname method:

x = Person("John", "Doe")
x.printname()
```




To keep the inheritance of the parent's __init__() function, add a call to the parent's __init__() function:

```python
class Person:
  def __init__(self, fname, lname):
    self.firstname = fname
    self.lastname = lname

  def printname(self):
    print(self.firstname, self.lastname)

class Student(Person):
  def __init__(self, fname, lname):
    Person.__init__(self, fname, lname)

x = Student("Mike", "Olsen")
x.printname()
```

```python
class arista:
    def __init__(self, lacp, lldp):
        self.lacp = lacp 
        self.lldp = lldp 

    def ocs(self):
        print(self.lacp , self.lldp)

class macsec(arista):
    def __init__(self, port, channel):
        arista.__init__(self, port, channel)

bgp = macsec("isis", "level-1")
bgp.ocs()
```

```python
class Person:
  def __init__(self, fname, lname):
    self.firstname = fname
    self.lastname = lname

  def printname(self):
    print(self.firstname, self.lastname)

class Student(Person):
  def __init__(self, fname, lname):
    super().__init__(fname, lname)

x = Student("Mike", "Olsen")
x.printname()
```

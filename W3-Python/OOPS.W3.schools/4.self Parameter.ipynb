```python
class intent:
    def __init__(routing, childgateway, parentgateway):
        routing.childgateway = childgateway
        routing.parentgateway = parentgateway

    def routemaps(bastion):
        print("Azure networking is best and easier" + bastion.childgateway)


p1 = intent("Cisco and Juniper", 545)
p1.routemaps()
 
```

```python
class Car:
  def __init__(self, brand, model, year):
    self.brand = brand
    self.model = model
    self.year = year

  def display_info(self):
    print(f"{self.year} {self.brand} {self.model}")

car1 = Car("Toyota", "Corolla", 2020)
car1.display_info()
```

```python
class ip:
    def __init__(self, nsg, nva, vwan):
        self.nsg = nsg
        self.nva = nva 
        self.vwan = vwan 

    def eagleeye(self):
        print(f"Azure service insights and {self.nsg} {self.nva}  {self.vwan}")

kusto1 = ip("src_ip and the 4096 rules", "palo alto", 54151)
kusto1.eagleeye()
```

```python
class srcip:
    def __init__(self, destip, srcip, nexthop):
        self.destip = destip
        self.srcip = srcip
        self.nexthop = nexthop

    def jarvis(self):
        print(f"{self.destip} {self.srcip}  {self.nexthop}")

ava1 = srcip("192.168.5.1", "10.0.0.1", "Firewall" )
ava1.jarvis()
```

```python
class Person:
  def __init__(self, name):
    self.name = name

  def greet(self):
    return "Hello, " + self.name

  def welcome(self):
    message = self.greet()
    print(message + "! Welcome to our website.")

p1 = Person("Tobias")
p1.welcome()
```

```python
class primary:
    def __init__(self, primary):
        self.primary = primary 

    def public(self):
        return "Public IP addres are routable in internet" + self.primary
    
    def private(self):
        ipaddress = self.public()
        print(ipaddress + "class A and Class B and Class C private address are used for rpouting")

j1 = primary("       scaling of firewall           ")
j1.private()
```

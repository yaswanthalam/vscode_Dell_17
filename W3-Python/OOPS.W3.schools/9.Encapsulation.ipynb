```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.__age = age  # Private property

p1 = Person("Emil", 25)
print(p1.name)
print(p1.__age)  # This will cause an error
```

Above is failing because private properties can not be accessed directly outside of the class that's the reason above code was resulting an error

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.__age = age

  def get_age(self):
    return self.__age

p1 = Person("Tobias", 25)
print(p1.get_age())
```

__ using under score is considered as private and it will not allow us to change or modify or access the values directly outside of the class

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.__age = age

  def get_age(self):
    return self.__age

  def set_age(self, age):
    if age > 0:
      self.__age = age
    else:
      print("Age must be positive")

p1 = Person("Tobias", 25)
print(p1.get_age())

p1.set_age(26)
print(p1.get_age())
```

```python
class vwan:
    def __init__(self, hub, gateway):
        self.hub = hub
        self.__gateway = gateway 

    def nva(self):
        return self.__gateway
    
    def firewall(self, gateway):
        if gateway > 0:
            self.__gateway = gateway

        else:
            print("VWAN is an SDWAN solution which is one of the best one")

p2s = vwan("S2Sgateway" , "P2S gateway")
print(p2s.nva())

p2s.firewall(626)
print(p2s.nva())

p2s.firewall(0)
print(p2s.nva())
```

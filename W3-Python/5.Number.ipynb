Python Numbers
There are three numeric types in Python:

int
float
complex

```python
x=1
y=2.5
z=1j

print(x)
print(y)
print(z)

print(type(x))
print(type(y))
print(type(z))
```

```python
x=1.0
y=2.5
z=99999

print(x)
print(y)
print(z)

print(type(x))
print(type(y))
print(type(z))
```

The sciendtific numbers that end includes E and e as exponential also behaves like an floating numbers

```python
x = 35e3
y = 12E4
z = -87.7e100

print(type(x))
print(type(y))
print(type(z))
```

```python
x = 3+5j
y = 5j
z = -5j

print(x)
print(y)
print(z)

print(type(x))
print(type(y))
print(type(z))
```

Type Conversion

```python
x=1
y=2.5
z=8j
u=88


a=float(x)
b=int(y)
c=complex(z)
d=complex(u)

print(a)
print(b)
print(c)
print(d)
```

Note: You cannot convert complex numbers into another number type.

```python
x=1
y=2.5
z=8j
u=88+8j


a=float(x)
b=int(y)
c=int(z)
d=float(u)

print(a)
print(b)
print(c)
print(d)
```

Random Number
Python does not have a random() function to make a random number, but Python has a built-in module called random that can be used to make random numbers:

```python
import random

print(random.randrange(1,10))
```

```python
import random 

for i in range (1,10):
    print(random.randrange(i))
```

```python
import random 

n=100
for i in range(n):
    print(random.randrange(i+1))
```

```python
import random 

n=100
for i in range(n):
    print(random.randrange(i))
```
